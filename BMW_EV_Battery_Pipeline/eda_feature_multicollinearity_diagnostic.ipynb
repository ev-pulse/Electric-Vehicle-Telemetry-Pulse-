```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 파일 경로 지정 및 데이터 로드
file_path = r'C:\Users\EL095\Microsoft\20260515\ProjectDataset\Combined_0.1_Dataset.csv'
df = pd.read_csv(file_path)

# (필요 시) 데이터 로드 확인
print(f"데이터 로드 완료: {df.shape[0]}행, {df.shape[1]}열")

# 2. 피처 컬럼 정의 (이미 정의하신 함수와 리스트 사용)
# base_exclude_cols, leakage_risk_cols는 앞서 정의하신 내용을 그대로 사용하시면 됩니다.
# 여기서는 예시로 no_leakage 설정의 피처를 추출합니다.
feature_cols = get_feature_columns(df, EXPERIMENTS['no_leakage']['exclude_cols'])

# 3. 상관관계 계산
corr_matrix = df[feature_cols].corr()

# 4. 0.85 이상의 높은 상관관계를 가진 쌍 추출
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_pairs = []
for col in upper_tri.columns:
    high_corr_features = upper_tri[col][upper_tri[col].abs() > 0.85]
    for row, val in high_corr_features.items():
        high_corr_pairs.append({'Feature 1': row, 'Feature 2': col, 'Correlation': round(val, 4)})

high_corr_df = pd.DataFrame(high_corr_pairs)

if not high_corr_df.empty:
    print("=== 상관관계가 0.85 이상인 피처 쌍 ===")
    display(high_corr_df.sort_values(by='Correlation', key=abs, ascending=False))
else:
    print("상관관계가 높은 피처 쌍이 없습니다.")

# 5. 히트맵 시각화
plt.figure(figsize=(18, 14))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, cmap='RdBu_r', vmax=1.0, vmin=-1.0, center=0, annot=False)
plt.title("Feature Correlation Matrix", fontsize=15)
plt.show()
```
